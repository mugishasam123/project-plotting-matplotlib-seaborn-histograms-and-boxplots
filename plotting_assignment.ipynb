```python
plt.figure(figsize=(10, 6))
sns.histplot(df['co2'], bins=10, color='tomato', kde=True)
plt.title('c (bins=10)')
plt.xlabel('CO2 Emissions (g/mile)')
plt.ylabel('Number of Vehicles')
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 6))
sns.histplot(df['co2'], bins=30, color='seagreen', kde=True)
plt.title('Distribution of Carbon Dioxide Emissions (bins=30)')
plt.xlabel('CO2 Emissions (g/mile)')
plt.ylabel('Number of Vehicles')
plt.grid(True)
plt.show()
```

Distribution of CO₂ Emissions
The histogram above shows the distribution of carbon dioxide (CO₂) emissions for vehicles in the dataset, measured in grams per mile, with 30 bins for greater detail.

Most vehicles emit between 250 and 450 g/mile of CO₂, with the highest concentration around 350–400 g/mile. This suggests that the majority of vehicles are typical passenger cars and midsize vehicles from the 2013 market, which have moderate emissions levels.

The distribution is right-skewed, meaning there are a smaller number of vehicles with much higher emissions (above 500 g/mile). These outliers are likely large SUVs, powerful sports cars, or trucks with bigger engines and lower fuel efficiency. On the lower end, only a few vehicles emit less than 250 g/mile-these are likely hybrids or especially efficient compact cars.

Using more bins in the histogram helps reveal subtle features in the data, such as small clusters of low-emission vehicles and the spread of high-emission outliers.

In summary:

Most vehicles fall in the moderate emissions range (250–450 g/mile).

There are fewer vehicles with very high or very low CO₂ emissions, creating a long right tail in the distribution.

The data highlights the diversity in vehicle emissions and suggests that while cleaner vehicles exist, high-emission models are still present and could be targeted for emissions reduction efforts.







```python
plt.figure(figsize=(10, 6))
sns.histplot(df['displ'], bins=15, color='goldenrod', kde=True)
plt.title('Distribution of Engine Displacement')
plt.xlabel('Engine Displacement (L)')
plt.ylabel('Number of Vehicles')
plt.grid(True)
plt.show()
```

 Distribution of Engine Displacement
The histogram shows that most vehicles have engine displacements between 2.0 and 4.0 liters, with a pronounced peak at around 2.0–2.5 liters. This indicates that mid-sized engines are the most common in the dataset, likely reflecting the popularity of mainstream sedans, crossovers, and midsize SUVs that balance power and fuel efficiency.

There are fewer vehicles with very small engines (below 2.0L), which are typically found in compact or economy cars designed for maximum fuel efficiency. Similarly, vehicles with large engines (above 4.0L) are also less frequent. These larger engines are usually found in high-performance sports cars, luxury vehicles, or trucks, which prioritize power over efficiency.

The distribution is slightly right-skewed, with a long tail extending toward higher displacements up to 7.0 liters. This tail highlights the presence of a small number of vehicles with very large engines.

In summary:

The dataset is dominated by vehicles with mid-sized engines (2.0–4.0L).

Small and large engine vehicles are present but much less common.

This pattern suggests a market preference for vehicles that offer a compromise between performance and efficiency, with high-performance or ultra-efficient vehicles representing a smaller share of the market

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['comb'], bins=20, color='royalblue', kde=True)
plt.title('Distribution of Combined Fuel Efficiency')
plt.xlabel('Combined Fuel Efficiency (mpg)')
plt.ylabel('Number of Vehicles')
plt.grid(True)
plt.show()
```

 Distribution of Combined Fuel Efficiency
The histogram displays the distribution of combined fuel efficiency (mpg) for vehicles in the dataset. The distribution is right-skewed, with the majority of vehicles achieving between 18 and 30 mpg. The highest concentration is around 22 to 25 mpg, indicating that most vehicles are moderately fuel efficient.

A smaller number of vehicles achieve higher fuel efficiency (above 30 mpg), which creates a long tail on the right side of the plot. These right-tail vehicles are likely hybrids or compact cars with advanced technology. Conversely, very few vehicles have extremely low fuel efficiency (below 18 mpg), which are likely high-performance sports cars or large trucks.

This distribution suggests:

Most vehicles in the dataset are typical passenger cars with moderate fuel efficiency.

High-efficiency vehicles are present but less common, reflecting gradual adoption of fuel-saving technologies.

The right-skewed shape highlights opportunities for increasing the average fuel efficiency through technology and policy.

```python

numeric_df = df.select_dtypes(include=[np.number])

corr_matrix = numeric_df.corr()

plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap of Numeric Features')
plt.tight_layout()
plt.show()
```

Heatmap Interpretation
The correlation heatmap shows strong relationships among several numeric features in the dataset:

Redundant Features:

feScore and ghgScore are almost perfectly correlated (correlation ≈ 1), so keeping both in a predictive model is unnecessary.

The fuel efficiency columns (comb, city, UCity, highway, UHighway) are all highly correlated with each other and with the environmental scores. Keeping just one (such as comb) is sufficient.

Negative Correlations:

co2 is strongly negatively correlated with fuel efficiency and environmental scores (correlations around -0.9 to -0.94), meaning vehicles with higher fuel efficiency have lower CO₂ emissions.

Engine size (displ) and number of cylinders are also negatively correlated with ghgScore, indicating that larger engines and more cylinders generally result in worse greenhouse gas scores.

Weak or Unrelated Features:

Columns like pv2 and pv4 have weak correlations with most other variables, suggesting they may not contribute much to predicting ghgScore.

id and year are identifiers or nearly constant and are not useful for prediction.
